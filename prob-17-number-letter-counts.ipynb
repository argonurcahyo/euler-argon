```python
ONES = ["",
        "one",
        "two",
        "three",
        "four",
        "five",
        "six",
        "seven",
        "eight",
        "nine",
        "ten",
        "eleven",
        "twelve",
        "thirteen",
        "fourteen",
        "fifteen",
        "sixteen",
        "seventeen",
        "eighteen",
        "nineteen"]
TENS = ["","",
        "twenty","thirty","forty","fifty","sixty","seventy","eighty","ninety"]
```

```python
def num_to_en(n):
    if 0 <= n < 20:
        return ONES[n]
    elif 20 <= n < 100:
        return f"{TENS[n//10]}{num_to_en(n % 10)}"
    elif 100 <= n < 1000:
        return f"{ONES[n//100]}hundred" + (f"and{num_to_en(n % 100)}" if n%100 !=0 else "")
    elif 1000 <= n < 10000:
        return f"{num_to_en(n//1000)}thousand{num_to_en(n % 1000)}"
```

```python
total = 0
for i in range(1,1001):
    en = num_to_en(i)
    total += len(en.replace(' ','')) 
    print(f"{en} | {total}")
    
ans = sum(len(num_to_en(i)) for i in range(1, 1001))
```
